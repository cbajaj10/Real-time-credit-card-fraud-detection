```python
appli_train = appli_train.reset_index(drop = True)
```

```python
appli_train['randNum'] = np.random.randint(1, 1000000, appli_train.shape[0])
appli_train['FraudLabel']=appli_train['fraud_label']
# Create a fraud column and random number column for comparison
fraud = appli_train['fraud_label']
appli_train.drop(labels=['fraud_label'], axis=1,inplace = True)
appli_train.insert(0, 'fraud_label', fraud)
appli_train.dtypes
```

## Calculate FDR Table for all the candidate variables

```python
appli_train.isnull().any()
appli_train_2 = appli_train.copy()
```

```python
Three_Percent = round(0.03 * len(appli_train))
TotalFraud = appli_train['fraud_label'].sum()
FdrTable = pd.DataFrame(columns=['Variable', 'FDR'])
appli_train_2['fraud_label'] = appli_train_2['fraud_label'].astype('float64')
for column in appli_train_2.iloc[:,2:]:
    FraudCaught = appli_train_2[[column,'fraud_label']].sort_values(by = [column], ascending = False)
    FDR = FraudCaught.iloc[0:Three_Percent,1:2].sum() / TotalFraud
    FdrTable = FdrTable.append({'Variable': column, 'FDR':float(FDR)}, ignore_index=True)


FdrTable = FdrTable.sort_values(by = 'FDR', ascending = False)
FdrTable
```

## Calculate KS for each candidate variable

```python
appli_train_3 = appli_train.copy()
KsTable = pd.DataFrame(columns=['Variable', 'KS'])
for column in appli_train_3.iloc[:,2:]:
    KS = sps.ks_2samp(appli_train_3.loc[appli_train_3['fraud_label'] == 0, column], appli_train_3.loc[appli_train_3['fraud_label'] == 1, column])
    KsTable = KsTable.append({'Variable': column, 'KS':KS[0] }, ignore_index=True)
KsTable = KsTable.sort_values(by = 'KS', ascending = False)
KsTable
```

## Compile two tables and Save to an excel

```python
appli_final=KsTable.merge(FdrTable,on='Variable')
```
